# Which are the sentences where "love" was used?

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
len(findings)
```

# What are the most used words?

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings
```

```python
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d
```

```python
d_list = [(value, key) for key, value in d.items()]
```

```python
d_list
```

```python
d_list_sorted = sorted(d_list, reverse=True)
```

```python
d_list_sorted
```

# Paragraphs with "love"

# Sentiment Analysis: the most positive and the most negative chapters.

```python
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download("vader_lexicon")
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer.polarity_scores("Hey, look how beautiful the trees are. I am gonna burn them.")
```

```python
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]
```

```python
for chapter in chapters:
    scores = analyzer.polarity_scores(chapter)
    print(scores)
```
